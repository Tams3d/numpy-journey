# Advanced Operation


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array(
    [  # 0     1      2       3       4
        [1, 150000, 180000, 220000, 250000],  # Paradise Biryani    0
        [2, 120000, 140000, 160000, 190000],  # Beijing Bites       1
        [3, 200000, 230000, 260000, 300000],  # Pizza Hub           2
        [4, 180000, 210000, 240000, 270000],  # Burger Point        3
        [5, 160000, 185000, 205000, 230000],  # Chai Point          4
    ]
)

print("Data Shape:", sales_data.shape)
print("Data of First three restaurants:\n", sales_data[:, 1:])
```

```python
# Total yearly sales
total_yearly_sales = np.sum(sales_data[:, 1:], axis=0)  # axis = 0 for Column-wise sum
print("Total sales of all restaurant per year:", total_yearly_sales)

total_restaurant_sales = np.sum(
    sales_data[:, 1:], axis=1
)  # axis = 01 for Column-wise sum

print("Total sales per restaurant in all years:", total_restaurant_sales)
```

```python
# Minimum sales per restaurant in all years
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales:", min_sales)
```

```python
# Maximum sales per year in all restaurant
max_yearly_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales in a year:", max_yearly_sales)
```

```python
# Average sales per restaurant in all years
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales of Restaurant:", avg_sales)
```

```python
years = np.arange(2021, 2025)
plt.plot(years, sales_data[1, 1:])
plt.plot(years, sales_data[2, 1:])
plt.plot(years, sales_data[3, 1:])
plt.plot(years, sales_data[4, 1:])
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Restaurant Sales")
plt.show()
```
