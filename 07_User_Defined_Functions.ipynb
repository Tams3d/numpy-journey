#### Sigmoid Function:

$$
\sigma(x) = \frac{1}{1 + e^{-x}} = \frac{e^x}{1 + e^x} = 1 - \sigma(-x)
$$


```python
# sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
array = np.linspace(-10, 10, 50)
sigmoid_array = sigmoid(array)
print("Array:", array)
print("\nSigmoid of array:", sigmoid_array)
```

```python
plt.plot(array, sigmoid_array)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.show()
```

#### Mean Squared Error:

$$
\text{Mean Squared Error (MSE)} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 \\[1em]

\begin{aligned}
\text{where:} \\
n &= \text{Total number of data points (samples)} \\
i &= \text{Index of a data point, such that } i = 1, 2, \dots, n \\
y_i &= \text{Actual (true) value of the } i^\text{th} \text{ sample} \\
\hat{y}_i &= \text{Predicted value of the } i^\text{th} \text{ sample} \\
(y_i - \hat{y}_i)^2 &= \text{Squared error for the } i^\text{th} \text{ sample} \\
\end{aligned}


$$


```python
# mean squared error
def mse(actual, predicted):
    return np.mean((actual - predicted) ** 2)
```

```python
actual = np.random.randint(2, 20, 10)
predicted = np.random.randint(2, 20, 10)
mse_value = mse(actual, predicted)

print("Actual:", actual)
print("Predicted:", predicted)
print("Mean Squared Error:", mse_value)
```

#### Binary Cross Entropy:

$$
L_{BCE} = -\frac{1}{n} \sum_{i=1}^{n} \left( y_i \cdot \log(\hat{y}_i) + (1 - y_i) \cdot \log(1 - \hat{y}_i) \right)

\\[2ex]

\begin{aligned}
\text{where:} \\
\text{n} &= \text{Total number of data points (samples)} \\
i &= \text{Index of a data point, } i = 1, 2, \dots, n \\
\textcolor{lime}{y_i} &= \text{Actual (true) label of the } i^{th} \text{ sample} \\
\textcolor{red}{\hat{y}_i} &= \text{Predicted value (probability) for the } i^{th} \text{ sample}
\end{aligned}

\\[2ex]

\underbrace{
\textcolor{lime}{y_i} \cdot \log\left(\textcolor{red}{\hat{y}_i}\right)
}_{\text{error for positive case}}
\quad
\underbrace{
\left(1 - \textcolor{lime}{y_i}\right) \cdot \log\left(1 - \textcolor{red}{\hat{y}_i}\right)
}_{\text{error for negative case}}
$$


```python
# Binary Cross Entropy


def bce(actual, predicted, eps=1e-15):
    predicted = np.clip(predicted, eps, 1 - eps)
    return -np.mean(actual * np.log(predicted) + (1 - actual) * np.log(1 - predicted))
```

```python
bce_value = bce(actual, predicted)

print("Actual:", actual)
print("Predicted:", predicted)
print("Binary Cross Entropy Loss:", bce_value)
```

### Plotting Graphs


```python
# y = x
x = np.arange(100)
y = x

plt.plot(x, y)
plt.title("y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
# y = x**2
x = np.linspace(-10, 10, 50)
y = x**2

plt.plot(x, y)
plt.title("y = x**2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
# sin(x) and cos(x)
# Degrees from -180 to 180
x_deg = np.linspace(-180, 180, 1000)
x_rad = np.deg2rad(x_deg)  # Convert degrees to radians

y = np.sin(x_rad)
z = np.cos(x_rad)

plt.plot(x_deg, y, label="sin(x)")
plt.plot(x_deg, z, label="cos(x)")


plt.xlabel("x (in degrees)")
plt.ylabel("y")
plt.grid()

plt.legend()
plt.xticks(np.arange(-180, 181, 45))
plt.show()
```

```python
# tan(x)
x_deg = np.linspace(-180, 180, 1000)
x_rad = np.deg2rad(x_deg)

y = np.tan(x_rad)
y[np.abs(y) > 10] = np.nan  # Remove extreme values near asymptotes

plt.plot(x_deg, y, label="tan(x)", color="orange")

plt.xlabel("x (in degrees)")

plt.ylabel("y")
plt.grid()
plt.legend()
plt.xticks(np.arange(-180, 181, 45))
plt.ylim(-10, 10)
plt.title("tan(x)")

plt.show()
```
